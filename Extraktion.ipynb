
Beim Ausf√ºhren sehe ich, dass einige Spalten fehlen: 'HISEI', 'IC010Q12HA', 'IC004Q01HA', 'IC010Q11HA', 'IC002Q01HA'. Ich suche die Codeb√ºcher durch und stelle fest, dass die Spalten aus dem Jahre 2015 entweder anders genannt  wurden ('HISEI', 'IC004Q01HA', 'IC002Q01HA') oder fehlen ('IC010Q11HA','IC010Q12HA'). Bez√ºglich des Nutzens von digitalen Ger√§ten im Jahre 2015 wurde noch zus√§tzlich das Alter beim ersten Zugang zum Computer abgefragt:

'IC003Q01TA': 'How old were you when you first used a computer?'

was aus meiner Sicht auch zu digitalen Ger√§ten z√§hlen kann. 

```python
columns_list_15 = columns_list.copy()
columns_list_15.remove('HISEI')  # l√∂sche die Spaltennamen, die nicht in Daten von 2015 sind
columns_list_15.remove('IC010Q12HA')
columns_list_15.remove('IC004Q01HA')
columns_list_15.remove('IC010Q11HA')
columns_list_15.remove('IC002Q01HA')

extend_list = ['hisei', 'IC002Q01NA', 'IC003Q01TA', 'IC004Q01TA'] # die Spaltennamen, wie sie in Daten 2015 sind
columns_list_15.extend(extend_list) # f√ºge die fehlenden Spaltennamen hinzu
#columns_list_15   
```

```python
df_kurz_15 = df_15[columns_list_15]
df_kurz_15
```

```python
df_kurz_15.to_csv('pisa_it_15.csv') # speichere der gek√ºrzter DataFrame als csv-Datei zur sp√§teren Verwendung 
```
