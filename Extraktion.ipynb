##### Daten von 2015
Links sind in Readme-Datei.

```python
df_15, meta_15 = pyreadstat.read_sav('CY6_MS_CMB_STU_QQQ.sav')
```

```python
df_15
```

```python
columns_15 = meta_15.column_names_to_labels
columns_15
```

```python
"""Ich möchte die gleichen Spalten wie im Jahre 2018 ausselektieren"""
df_kurz_15 = df_15[columns_list]
df_kurz_15
```


Ich stelle fest, dass einige Spalten fehlen: 'HISEI', 'IC010Q12HA', 'IC004Q01HA', 'IC010Q11HA', 'IC002Q01HA'. Ich suche die Codebücher durch und stelle fest, dass die Spalten aus dem Jahre 2015  wurden entweder anders genannt ('HISEI', 'IC004Q01HA', 'IC002Q01HA') oder fehlen ('IC010Q11HA','IC010Q12HA'). Bezüglich des Nutzens von digitalen Geräten im Jahre 2015 wurde noch zusätzlich das Alter beim ersten Zugang zum Computer abgefragt:

'IC003Q01TA': 'How old were you when you first used a computer?'

was aus meiner Sicht auch zu digitalen Geräten zählen kann. 

```python
columns_list_15 = columns_list.copy()
columns_list_15.remove('HISEI')  # lösche die Spaltennamen, die nicht in Daten von 2015 sind
columns_list_15.remove('IC010Q12HA')
columns_list_15.remove('IC004Q01HA')
columns_list_15.remove('IC010Q11HA')
columns_list_15.remove('IC002Q01HA')

extend_list = ['hisei', 'IC002Q01NA', 'IC003Q01TA', 'IC004Q01TA'] # die Spaltennamen, wie sie in Daten 2015 sind
columns_list_15.extend(extend_list) # füge die fehlenden Spaltennamen hinzu
columns_list_15 
```

```python
df_kurz_15 = df_15[columns_list_15]
df_kurz_15
```

```python
df_kurz_15.to_csv('pisa_it_15.csv') # speichere der gekürzter DataFrame als csv-Datei für spätere Nutzung 
```
